```python
url='https://supermercado.eroski.es/es/supermercado/2059806-alimentacion/'
pag=urllib.request.urlopen(url)
doc=BeautifulSoup(pag)
enlaces_l=[]
enlaces_final=[]
cnt=0
for i in doc.find_all("li"):
    if i.find("a",class_="not_traceable")!=None: 
        #print(i.find("a",class_="not_traceable").get("href"))
        enlaces_l.append(i.find("a",class_="not_traceable").get("href"))
        enlaces_final.append("/".join(enlaces_l[cnt].split("/")[:-2]))
        cnt=cnt+1
enlaces_l = list(dict.fromkeys(enlaces_l))
enlaces_final = list(dict.fromkeys(enlaces_final))
```
